# segmentation

```python
img = cv2.imread(r'tulips.jpg')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

lr = np.array([0,120,70])
ur = np.array([8,255,255])
mask1 = cv2.inRange(hsv,lr,ur)

lr = np.array([170,120,70])
ur = np.array([180,255,255])
mask2 = cv2.inRange(hsv,lr,ur)

mask = mask1 | mask2

res = cv2.bitwise_and(img,img,mask=mask)

plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```

```python
img = cv2.imread(r'tulips.jpg')

ly = np.array([20, 70, 70])
uy = np.array([30, 255, 255])
mask=cv2.inRange(hsv,ly,uy)

res=cv2.bitwise_and(img,img,mask=mask)

plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```
